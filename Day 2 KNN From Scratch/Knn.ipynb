```python
# implementation of KNN from Scratch
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
import math
def euclidean_distance(p1,p2):
    dist=[]
    d=[]
    for i in range(len(p2)):
        a=(p1[i]-p2[i])**2
        d.append(a)
    dist.append(math.sqrt(np.sum(d)))
    return(dist)
```

```python
neighbors=16
dummy=[]
distances=[]
distance_label= []
y_pred=[]
dist_list=[]
for j in range(len(X_test)):
    dist_list=[]
    for k in range(len(X_train)):
        dist_list.append(euclidean_distance(X_test[j], X_train[k]))
   
    for l,m in enumerate(dist_list):
        dummy.append(l)
        dummy.append(m)
        distances.append(dummy)
        dummy=[]
        
    #Sorting to get the nearest points to the test_example
    
    distances.sort(key= lambda x:x[1])
    
    #pick only neighbors
    distances= distances[:neighbors]
    
    for n in distances:
        distance_label.append(y_train[n[0]])
    
    # Voting
    cnt= Counter()
    for i in distance_label:
        cnt[i]+=1
    # Pick the Label with highest Votes
    for p in cnt.most_common(1):
        y_pred.append(p[0])
    
    dist_list=[]
    distances=[]
    distance_label=[] 
```

```python
c=0
for i in range(len(y_pred)):
    if y_pred[i]==y_test[i]:
        c+=1
print("Accuracy is :",(c/len(y_pred)*100), "%")
```
